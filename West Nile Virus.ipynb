```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
weather = pd.read_csv("raw_data/weather.csv")
print(weather.shape)
weather.head()
```

```python
weather.isna().sum()
```

```python
weather['Date']=pd.to_datetime(weather['Date'])
weather.info()
```

```python
weather_join = weather.copy()
weather_join = weather_join.drop(columns=['Depth','Water1'])
weather_join.head()
```

```python
for index, value in enumerate(weather_join['Depart']):
    if index%2!=0:
        weather_join.loc[index,'Depart'] = weather_join.loc[index-1,'Depart']
        weather_join.loc[index,'SnowFall'] = weather_join.loc[index-1,'SnowFall']
```

```python
weather_join.loc[:,'CodeSum'] = weather_join.loc[:,'CodeSum'].str.split()
```

```python
total = []
for value in weather_join['CodeSum']:
    for items in value:
        if items not in total:
            total.append(items)
total
```

```python
weather_join[total]=0
for index,value in enumerate(weather_join['CodeSum']):
    for items in value:
        weather_join.loc[index,items]=1
weather_join = weather_join.drop(columns=['CodeSum'])
```

```python
weather_join
```

```python
for index, value in enumerate(weather_join['Tavg']):
    if value=="M":
        weather_join.loc[index,'Tavg']=(weather_join.loc[index,'Tmax']+weather_join.loc[index,'Tmin'])/2
        
weather_join
```

```python
weather_join = weather_join.astype({'Tavg': 'int64'})
weather_join.describe()
```

```python
train = pd.read_csv("raw_data/train.csv")
print(train.shape)
train.head()
```

```python
train.isna().sum()
```

```python
train['Date']=pd.to_datetime(train['Date'])
train.info()
```

```python
train_join = train.copy()
train_join = train_join.drop(columns=['Address','Block','Street','AddressNumberAndStreet'])
train_join.loc[:,'Species'] = train_join.loc[:,'Species'].str.split(pat='/')
train_join.head()
```

```python
total = []
for value in train_join['Species']:
    for items in value:
        if items not in total:
            total.append(items)
            
for index, value in enumerate(total):
    if "CULEX " in value:
        total[index]= value[6:]

total.pop(1)
total
```

```python
train_join[total]=0
for index,value in enumerate(train_join['Species']):
    for items in value:
        for mosq_type in total:
            if mosq_type in items:
                train_join.loc[index,mosq_type]=1
train_join = train_join.drop(columns=['Species','Trap'])
train_join
```
